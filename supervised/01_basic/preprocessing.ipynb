```python
import pandas as pd
import numpy as np
import sklearn
import os
```

```python
data = {
    'size': ['XL', 'L', 'M', 'L', 'M'],
    'color': ['red', 'green', 'blue', 'green', 'red'],
    'gender': ['female', 'male', 'male', 'female', 'female'],
    'price': [199.0, 89.0, 99.0, 129.0, 79.0],
    'weight': [500, 450, 300, 380, 410],
    'bought': ['yes', 'no', 'yes', 'no', 'yes']
}

df_raw = pd.DataFrame(data=data)
df_raw
```

```python
df = df_raw.copy()
df
```

```python
for col in ['size', 'color', 'gender', 'bought']:
    df[col] = df[col].astype('category')
df['weight'] = df['weight'].astype('float')
df.info()  
```

###### Przygotowanie danych do klasyfikacji LabelEncoder służy do binaryzowania zmiennej objaśnianej - TARGET

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit_transform(df['bought'])
```

```python
df['bought'] = le.fit_transform(df['bought'])
df
```

##### Funkcja get dummies załatwia wszytko za jednym razem musiemy jednakże pamiętać o drop_first = True !!! 

```python
df_anal = pd.get_dummies(data = df_raw, drop_first='True')
df_anal
```

##### Skalowanie danych w bibliotece pandas std() jest estymatorem nie obciązonym 

```python
from sklearn.preprocessing import StandardScaler

scalar = StandardScaler()
df_anal[['price', 'weight']] = scalar.fit_transform(df_anal[['price','weight']])
df_new = df_anal.copy()
```
